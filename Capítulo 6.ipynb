```python
from matplotlib import pyplot as plt
import pandas as pd
```

##### 1. Por meio do _dataset_ `space.csv`, trace um gráfico em barras mostrando quantas empresas espaciais os EUA e a China possuem.

Dica: após realizar os devidos condicionais no dataset para cada um dos países, utilize o método `unique()` para retirar resultados repetidos.


```python
df1 = pd.read_csv('data/space.csv', delimiter=';')
df1.head()
```

```python
usa_companies = df1['Company Name'].where(df1['Location'].str.find('USA') != -1).dropna().unique()
usa_companies
```

```python
china_companies = df1['Company Name'].where(df1['Location'].str.find('China') != -1).dropna().unique()
china_companies
```

```python
plt.title('Space Companies Owned by Countries')
plt.bar('USA', len(usa_companies), color='blue')
plt.bar('China', len(china_companies), color='red')
plt.grid(True)
plt.show()
```

##### 2. Por meio do _dataset_ `paises.csv`, trace dois gráficos de linhas em um mesmo plano cartesiano, um mostrando a taxa de mortalidade (`Deathrate`) e outro a taxa de natalidade (`Birthrate`) dos países da América do Norte (`NORTHERN AMERICA`)

Dica: após realizar o condicional no _dataset_ para pegar os países da América do Norte, você já poderá chamar o método plot, passando como argumento para ele slicings sobre dataset resultante do condicional.


```python
df2 = pd.read_csv('data/paises.csv', delimiter=';')
df2.head()
```

```python
na_countries = df2.where(df2['Region'].str.find('NORTHERN AMERICA') != -1).dropna()
na_countries
```

```python
plt.title('Birthrate in the Northern America')
plt.plot(na_countries['Country'], na_countries['Birthrate'], color='blue')
plt.xticks(rotation=30)
plt.grid(True)
plt.show()
```

```python
plt.title('Deathrate in the Northern America')
plt.plot(na_countries['Country'], na_countries['Deathrate'], color='red')
plt.xticks(rotation=30)
plt.grid(True)
plt.show()
```
